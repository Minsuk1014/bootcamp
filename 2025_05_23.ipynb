```python
import re
import os
import requests 
from bs4 import BeautifulSoup # html passing을 도와줌.
from urllib.parse import urlparse, urlencode, quote 
import pathlib

def my_request(url, method = "get"):
        head = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36 Edg/136.0.0.0'}
        if method == "get" :
            return requests.get(url,headers=head) # r.text -> 단순하게 한줄로 텍스트로 나열
```

```python
def news(path = "news"):
    bs = my_request("https://news.kbs.co.kr/news/pc/view/view.do?ncd=8261268")
    bs = BeautifulSoup(bs.text)
    title = bs.find('h4',class_="headline-title").text
    detile = bs.find('div',class_='detail-body font-size').text
    Dtime=bs.find('div',class_="dates").text
    content = f"{title}\n{Dtime} \n {detile}"
    pathlib.Path(f"./news/{path}").mkdir(parents=True, exist_ok=True)
    f = open(f"./news/{path}/{title}.txt", 'w', encoding='utf-8')
    f.write(content.strip())
    f.close

news("news")







```
