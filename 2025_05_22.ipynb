```python
# 1단계 단순 함수 만들기

import os

import requests 

url = "https://www.melon.com/song/detail.htm?songId=1698598"
r = requests.get(url)

head = {'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/136.0.0.0 Safari/537.36 Edg/136.0.0.0'}
# 정보를 가져오기 위해서, 서버가 봇을 차단하는 걸 방지하기 위해서 사용.
r = requests.get(url, headers=head)

r.status_code # 현재 상태 값

from bs4 import BeautifulSoup # html passing을 도와줌.
bs = BeautifulSoup(r.text)

lyrics = BeautifulSoup(str(bs.find("div", id="d_video_summary")).replace("<br/>","\n")).text # 구체적인 id값이 있을 때,
# 다음과 같이 형 변환해서, 필요한 내장 함수를 사용한후 다시 바꿔줄 수 있음.


def song_write():
    title = bs.find("div", class_="song_name").text.replace("곡명","").strip() # strip() -> 양끝제거
    atist = bs.find("div", class_="artist").text.replace("곡명","").strip() 
    if not os.path.isdir("./lyrics"):
        os.mkdir("./lyrics")
    f = open(f"./lyrics/{atist}-{title}.txt", "w", encoding = "utf-8") # 파일을 저장 할 수도 있고, 열수도 있고.. 

    f.write(lyrics.strip())
    f.close

song_write()
```
